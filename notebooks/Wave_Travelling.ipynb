```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
```

```python
def animate(data):
    x = np.linspace(-10,10,nu)
    y = data
    line.set_data(x,y)
    return line,
```

```python
def step(nu):
    temp = np.ones(nu)
    temp[:(nu)*2./4.] = 0
    return temp
```

```python
def delta(nu,dx):
    temp = np.zeros(nu)
    temp[(nu)*0.5] = 1/dx
    return temp
```

```python
#Basic initial condition parameters
#defining grid size, time steps, CFL condition, etc...
nt = 20 # Time duration
h = 0.01 # Cell size
nu = 1 + 4.0/h # Number of points in space

c = 0.5 # Wave speed

x = np.linspace(-10,10,nu) # To avoid the effect at the end
u = step(nu)
plt.plot(x, u)
plt.xlim(-2,2)
plt.ylim(-0.5,1.5);
```

# Explicit Method - Forward Euler

```python
def forward_euler(u, nt, dt, r): # Forward-Euler for both space and time
    #initialize our results array with dimensions nt by nx
    u_n = np.zeros((nt,len(u)))      
    #copy the initial u array into each row of our new array
    u_n[:,:] = u.copy()              
    
    A = np.zeros((len(u), len(u)))
    i,j = np.indices(A.shape)
    A[i==j] = 1
    A[i==j-1] = 0.5*r
    A[i==j+1] = -0.5*r
    b = np.zeros(len(u))
    b[0] = -0.5*r
    b[-1] = 0.5*r
    
    for t in range(1,nt):
        u_n[t,:] = A.dot(u)
        u = u_n[t].copy()
        
    return u_n
```

$$
u_{j}^{n+1} = u_{j}^{n} + c\frac{1}{2} \frac{\Delta t}{\Delta x}(u_{j+1}^{n} - u^{n}_{j-1})
$$

```python
r = 1.0 
dt = r*h/c

u = step(nu)      #make sure that u is set to our expected initial conditions
#u = delta(nu, h)
u_n = forward_euler(u, nt, dt, r)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(-2,2),ylim=(-0.5,1.5));
line, = ax.plot([],[]);

anim = animation.FuncAnimation(fig, animate, frames=u_n, interval=50)
display_animation(anim, default_mode='once')
```

# Implicit Method - Backward Euler

```python
def backward_euler(u, nt, dt, r): # Forward-Euler for both space and time
    #initialize our results array with dimensions nt by nx
    u_n = np.zeros((nt,len(u)))      
    #copy the initial u array into each row of our new array
    u_n[:,:] = u.copy()              
    
    # Create the matrix
    A = np.zeros((len(u), len(u)))
    i,j = np.indices(A.shape)
    A[i==j] = 1
    A[i==j-1] = -0.5*r
    A[i==j+1] = 0.5*r
    b = np.zeros(len(u))
    b[0] = -0.5*r
    b[-1] = 0.5*r
    
    for t in range(1,nt):
        u_n[t,:] = np.linalg.inv(A).dot(u+b)
        u = u_n[t].copy()
        
    return u_n
```

$$
u_{j}^{n+1} = u_{j}^{n} + c\frac{1}{2} \frac{\Delta t}{\Delta x}(u_{j+1}^{n+1} - u^{n+1}_{j-1})
$$

```python
u = step(nu)      #make sure that u is set to our expected initial conditions
#u = delta(nu, h)
u_n = backward_euler(u, nt, dt, r)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(-2,2),ylim=(-0.5,1.5));
line, = ax.plot([],[]);

anim = animation.FuncAnimation(fig, animate, frames=u_n, interval=50)
display_animation(anim, default_mode='once')
```

# Implicit Method - Crank-Nicholson

$$
u_{j}^{n+1} = u_{j}^{n} + c\frac{1}{4} \frac{\Delta t}{\Delta x}(u_{j+1}^{n} - u^{n}_{j-1}) + c\frac{1}{4} \frac{\Delta t}{\Delta x}(u_{j+1}^{n+1} - u^{n+1}_{j-1})
$$

```python
def crank_nicholson(u, nt, dt, r): # Forward-Euler for both space and time
    #initialize our results array with dimensions nt by nx
    u_n = np.zeros((nt,len(u)))      
    #copy the initial u array into each row of our new array
    u_n[:,:] = u.copy()              
    
    # Create the matrix
    A1 = np.zeros((len(u), len(u)))
    A = np.zeros((len(u), len(u)))
    i,j = np.indices(A.shape)
    A1[i==j] = 1
    A1[i==j-1] = -0.25*r # There is a typo in the reference
    A1[i==j+1] = +0.25*r
    A = A1.T
    b = np.zeros(len(u))
    b[0] = -0.5*r
    b[-1] = 0.5*r
    
    for t in range(1,nt):
        u_n[t,:] = np.linalg.inv(A1).dot(A.dot(u)+b)
        u = u_n[t].copy()
        
    return u_n
```

```python
u = step(nu) #make sure that u is set to our expected initial conditions
#u = delta(nu, h)
u_n = crank_nicholson(u, nt, dt, r)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(-2,2),ylim=(-0.5,1.5));
line, = ax.plot([],[]);

anim = animation.FuncAnimation(fig, animate, frames=u_n, interval=50)
display_animation(anim, default_mode='once')
```
